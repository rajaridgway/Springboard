# Ridgway - Adult Income: Data Storytelling
An marketing team for a small technology start-up is interested in understanding in how to market their new devices to potential clients. Their market research has indicated that individuals making below 50,000 dollars a year a more likely to want their low-end device while those making above 50,000 dollars a year are more likely to want their high-end device. They have asked for insights from the Census income dataset ([more information](https://www.cs.toronto.edu/~delve/data/adult/adultDetail.html)) about factors that might impact their marketing strategies for both the low-end and high-end devices.

This notebook analyzing the 1994 Income dataset from the US Census to provide insights for the marketing team.

## Data Sources
- adult.csv: [Kaggle Adult Income data set](https://www.kaggle.com/wenruliu/adult-income-dataset/version/2)

## Attributes
There are 15 columns in the dataset:

| Column Name | Description | Variable Type | Data Type |
| :--- | :--- | :--- | :---: |
| age | Age of individual | cotinuous | int64 |
| workclass | Type of employment (e.g., private, government) | nominal | object |
| fnlwgt | Final weight | continuous | int64 |
| education | Highest level of education completed | nominal | object |
| educational-num | Years of education | continuous | int64 |
| marital-status | Martial status of individual | nominal | object |
| occupation | General occuptation of individual  | nominal | object |
| relationship | The individual's relationship in a family | nominal | object |
| race | Individual's race | nominal | object |
| gender | Individual's gender | nominal | object |
| capital-gain | Capital gains for the individual | continuous | int64 |
| capital-loss | Capital losses for the individual | continuous | int64 |
| hours-per-week | Number of hours worked by the individual per week | continuous | int64 |
| native-country | Individual's native country | nominal | object |
| income | Whether or not the individual makes more that $50k per year | nominal | object |

Note that details for the attributes were taken from the a [USCD report](https://cseweb.ucsd.edu/classes/sp15/cse190-c/reports/sp15/048.pdf). 

## Known Data Issues

### Missing Data
A total of 3620 records have at least one "?" value in workclass, occupation, or native-country. The "?" indicates a missing value. **To clean the data set, these records are removed.**
1. workclass: 2799 values are "?"
2. occupation: 2809 values are "?"
3. native-country: 857 values are "?"

## Import Libraries

```python
import pandas as pd
from pathlib import Path
from datetime import datetime
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Load Data

```python
df = pd.read_csv('adult.csv')
```

```python
df.info()
```

## Clean Data
1. Drop all rows with ? as one of the values

```python
# Remove all rows in the three columns that have "?" values
cols_missing = ['workclass', 'occupation', 'native-country']

for col in cols_missing:
    df = df[~df[col].isin(['?'])]
```

```python
# Determine new number of records
df.shape
```

## Exploratory Visualization
1. Countplots for each of the nominal variables
2. Histograms, boxplots, and correlation heatmap for each of the continuous variables

```python
# Define a function to plot countplots for nominal variables
def countplot(y, hue, xlabel, title):
    sns.set()
    fig, ax = plt.subplots(figsize=(15,10))
    ax = sns.countplot(y=y, data=df, hue=hue, order = df[y].value_counts().index)
    plt.xlabel(xlabel)
    plt.ylabel('Frequency')
    plt.title(title)
    sns.despine()
    plt.legend(title = 'Income Classification', bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    
    ax.bar_label(ax.containers[0], padding = 3)
    ax.bar_label(ax.containers[1], padding = 3)

    plt.show()
```

## Targeted Questions with Visuals
Given that the goal of this investigation is to identify which sub-groups to target for low- and high-end devices, the remainder of the notebook will consider the following questions:
1. Should high- or low-end devices be targeted by hours per week?
2. Should high- or low-end devices be targeted by age?
3. Should high- or low-end devices be targeted by education level?

```python
#Limit the dataframe
df = df[['age', 'education', 'educational-num','occupation', 
         'gender', 'income', 'hours-per-week']]
```

### Hours Per Week and Gender

```python
# Plot hours worked by income and disaggregate by gender
plt.figure(figsize=(20,10))
sns.boxplot(x='income',y ='hours-per-week', hue='gender',data=df)
plt.show()
```

<div class="alert alert-block alert-info">
When considering the number of hours worked per week, those who work more tend to make more money. <b>However, the differences are not large enough to consider hours worked per week as a differentiator for the marketing campaigns.</b>
</div>

### Age and Gender

```python
# Plot age worked by income and disaggregate by gender
plt.figure(figsize=(20,10))
sns.boxplot(x='income',y ='age', hue='gender',data=df)
plt.show()
```

```python
# Determine average age by income and gender
df.groupby(['income', 'gender'])['age'].mean()
```

<div class="alert alert-block alert-info">
When considering age, those who are older tend to make more money. This is true for both male and female respondents - the age difference is six years for female respondents and seven years for male respondents. <b>As a result, age should be considered as a differentiator for the campaign - younger (i.e., 20-30s) males and females for the low-end device and older (i.e., 40+) males and females for the high-end device.</b>
</div>

### Education and Gender

```python
# Plot years of education by income and disaggregate by gender
plt.figure(figsize=(20,10))
sns.boxplot(x='income',y ='educational-num', hue='gender',data=df)
plt.show()
```

```python
# Determine average age by income and gender
df.groupby(['income', 'gender'])['educational-num'].mean()
```

<div class="alert alert-block alert-info">
When considering education, those with more education tend to make more money. This is true for both male and female respondents - the years of education difference is about two years for female and male respondents.<b>As a result, age should be considered as a differentiator for the campaign - younger (i.e., high school completion) males and females for the low-end device and older (i.e., some college or college) males and females for the high-end device.</b>
</div>
